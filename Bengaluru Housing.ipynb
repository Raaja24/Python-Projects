```python
def isfloat(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df[~df['total_sqft'].apply(isfloat)].head(10)
```

```python
def convert_sqft_tonum(x):
    token=x.split('-')
    if len(token)==2:
        return (float(token[0])+float(token[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df =df.copy()
```

```python
df['total_sqft'] = df['total_sqft'].apply(convert_sqft_tonum)
```
